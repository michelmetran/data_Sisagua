O presente documento visa avaliar se o parâmetro de *cloro residual* dos sistemas de abastecimento público de um determinado município está em conformidade com os padrões estabelecidos no Anexo XX da Portaria de Consolidação MS nº 05/2017, bem como outros aspectos. Tal preocupação surgiu em meio a pandemia do COVID-19, quando então a Diretora Técnica do Centro de Vigilância Sanitária — órgão vinculado à Coordenadoria de Controle de Doenças da Secretaria de Saúde do Estado de São Paulo — publicou o Comunicado CVS-SAMA nº 6, de 23.03.2020 , trazendo a preocupação em relação a desinfeção e outras questões destacadas abaixo:

1.	Por conta disto, é necessária a oferta ininterrupta de água à população nos padrões de potabilidade estabelecidos pelas normas federal (em especial, o Anexo XX da Portaria de Consolidação 5 do Ministério da Saúde, de 03-10-2017 – antiga Portaria 2914/2011) e estadual (Resolução SS-65/2016);
2.	Neste contexto de crise sanitária, os responsáveis pelos sistemas e soluções alternativas de abastecimento devem, na medida de suas possibilidades, intensificar o controle da qualidade da água destinada ao consumo humano, de modo que os consumidores possam cumprir a contento a orientação dos órgãos de saúde para reforçar a higiene pessoal e do ambiente;
3.	Um dos aspectos fundamentais nesse momento é garantir as concentrações de cloro na água tratada de acordo com o padrão de potabilidade estabelecido em norma, bem como outros parâmetros que conferem potabilidade da água;
4.	Cabe destacar às autoridades sanitárias a importância de se reforçar a mensuração em campo dos teores de Cloro Residual Livre, de maneira a garantir a presença desse produto na rede de distribuição e nos pontos de abastecimento no teor determinado em norma;

Ainda, complementarmente, em meio a essa discussão, o Projeto Conexão Água, da Câmara de Meio Ambiente do Ministério Público Federal, publicou as “recomendações para prevenção do contágio da COVID-19 (novo coronavírus –SARS-CoV-2) pela água e por esgoto doméstico” , sendo pontuado o que segue:

1.	A Portaria de Consolidação MS n. 05/2017 (Anexo 4 do Anexo XX – Origem: PRT MS/GM n. 2914/2011) estabelece que, após a desinfecção, a água deve conter um teor mínimo de cloro residual livre de 0,5 mg/L, sendo obrigatória a manutenção de no mínimo 0,2 mg/L em qualquer ponto da rede de distribuição, recomendando-se que a cloração seja realizada em pH inferior a 8,0 e em tempo de contato mínimo de 30 minutos.
2.	É de grande importância que os responsáveis pelos sistemas de abastecimento público procedam a desinfecção da água antes da distribuição. A etapa de desinfecção da água tem caráter preventivo e corretivo. No entanto, há, ainda, muitos municípios brasileiros cujas águas de abastecimento público não são submetidas à etapa de desinfecção antes da distribuição.
3.	Às estações de tratamento de água que ainda não dispõem da etapa de cloração, recomenda-se a contratação de profissional técnico habilitado junto ao CREA/CRQ ou Conselho de classe competente, para orientar a inclusão dessa etapa de desinfecção, conforme as diretrizes normativas.
4.	É obrigatório que as águas de abastecimento público recebam o tratamento com a etapa de desinfecção. As águas provenientes de mananciais superficiais devem incluir os processos de filtração e desinfecção, cumprindo o que determina a Portaria de Consolidação MS nº 05/2017.

## Número de Amostras de Cloro Residual

As amostras de cloro residual são feitas pelo prestador do serviço de abastecimento público (controle), bem como pela vigilância sanitária municipal (vigilância), sendo esse último o principal foco destas análises.

De acordo com a [Diretriz Nacional do Plano de Amostragem da Vigilância da Qualidade da Água para Consumo Humano da Vigilância Sanitária](http://bvsms.saude.gov.br/bvs/publicacoes/diretriz_nacional_plano_amostragem_agua.pdf), o número mínimo de amostras mensais de cloro residual a ser realizada pela vigilância varia de acordo com a população, conforme tabela abaixo apresentada.

![Cloro](https://i.imgur.com/R0UdN6K.png)

Ainda, de acordo com as diretrizes, *o número mínimo mensal de análises previsto para o Plano de Amostragem Básico é definido em função das faixas populacionais e constitui um quantitativo único a ser distribuído para o monitoramento da qualidade da água referente às três formas de abastecimento de água (SAA, SAC e SAI)*. Portanto, para analisar se a vigilância sanitária municipal segue as diretrizes estipuladas pelo Ministério da Saúde para amostragem de cloro residual, não será feito, nesse momento, detalhamento sobre o número de amostras realizadas para cada sistema de abastecimento, avaliando-se somente o número total de amostras que devem ser realizadas pela vigilância municipal.

Nos códigos abaixo, a tabela que consta nas diretrizes do Ministério da Saúde foi codificada, possibilitando análises.

```python
# Criando tabela
df_cloro = pd.DataFrame({'População De': [0, 5001, 10001, 50001, 200001, 500001],
                         'População Até': [5000, 10000, 50000, 200000, 500000, np.inf],
                         'Nº de Amostras': [6, 9, '8 + (1 para cada 7,5 mil habitantes)', '10 + (1 para cada 10 mil habitantes)', '20 + (1 para cada 20 mil habitantes)', '35 + (1 para cada 50 mil habitantes)']
                        })
# Tabela
display(df_cloro)
```

Uma vez cofificada, o *campo* descritivo **Nº de Amostras**, que contem o número de análises, foi dividido entre o **Nº de Amostras Fixo** e **Nº de Amostras Variável**, possibilitando a automatização dos cálculos.

```python
# Criando e Ajustando Coluna Adicional
df_cloro['Nº de Amostras Fixo'] = df_cloro['Nº de Amostras'].str.split('+').str[0]
df_cloro['Nº de Amostras Fixo'] = df_cloro['Nº de Amostras Fixo'].fillna(df_cloro['Nº de Amostras']).astype(float)

# Criando e Ajustando Coluna Adicional
df_cloro['Nº de Amostras Variável'] = df_cloro['Nº de Amostras'].str.split('+').str[-1]
df_cloro['Nº de Amostras Variável'] = df_cloro['Nº de Amostras Variável'].fillna(0)
df_cloro['Nº de Amostras Variável'] = df_cloro['Nº de Amostras Variável'].replace({'\(1 para cada': '','mil habitantes\)': '',',': '.'}, regex=True).astype(float)
df_cloro['Nº de Amostras Variável'] = (df_cloro['Nº de Amostras Variável']*1000).astype(int)

# Tabela
display(df_cloro)
```

Com a tabela codificada, foi escrita uma função que calcula o número de amostras, seguindo as condições recomendadas pelo Ministério da Sáude.

```python
def numero_amostras(x, df_cloro):
    array = np.where((x >= df_cloro['População De']) & (x <= df_cloro['População Até']) & (df_cloro['Nº de Amostras Variável']>0),
                     (df_cloro['Nº de Amostras Fixo'] + x/(df_cloro['Nº de Amostras Variável'])),
                     
                     np.where((x >= df_cloro['População De']) & (x <= df_cloro['População Até']) & (df_cloro['Nº de Amostras Variável']==0),
                              df_cloro['Nº de Amostras Fixo'],np.nan))
    
    array = np.trunc(array)
    array = array[~np.isnan(array)]
    return array[0]
```

Abaixo são demonstrados alguns exemplos de cálculo empregando a tabela codificada e função. A definição do número mínimo de amostras pode ser feita utilizando um número isolado, ou em uma série de dados.

```python
# Usando a função em uma série
df_pop = pd.DataFrame({'População': [1500, 10000, 180000, 350000, 600000, 650000]})
df_pop['Nº Amostras'] = df_pop['População'].apply(lambda x: numero_amostras(x, df_cloro))
print('> Usando uma série de dados aleatórios:')
display(df_pop)

# Usando a função com apenas um número individual
população = 300000
n_amostras = numero_amostras(população, df_cloro)
print('> Usando um número de população isolado:')
print('Nº de Amostras de Cloro recomendadas para '+ str(população) + ' hab.: ' + str(n_amostras))
```

Visando compreender como se dá a distribuição do número de amostras em função do aumento da população, foi elaborado um gráfico que demonstra que o número de amostras tem um crescimento logaritmico, ou seja, para municípios com população até 200.000 hab há um aumento continuo do número de amostras, até 500.000 habitantes o aumento do número de amostras não segue a mesma tendância, sendo reduzido. Em municípios com mais de 500.000 o aumento do número de amostras é ainda menor.

```python
# Cria tabela com população
df = pd.Series(np.arange(1000, 1100000, 3000)).array
df = pd.DataFrame(df, columns=['População'])

# Ajustes da tabela
df['População'] = df['População'].astype(int)
df['Nº Amostras'] = df['População'].apply(lambda x: numero_amostras(x, df_cloro))

# Gráfico
fig, ax = plt.subplots()
df.plot(x ='População', y='Nº Amostras', ax=ax, legend=False)
ax.ticklabel_format(style='plain')
ax.set_xlabel('População')
ax.set_ylabel('Nº Amostras')
plt.show()
```

# Definição do Município

Os códigos dos municípios são definidos pelo IBGE, conforme se explica com mais detalhes nesse [*link*](https://www.ibge.gov.br/explica/codigos-dos-municipios.php). Atualmente, conforme se pode observar, o código do município tem 7 dígitos.

O código tem uma sistematização que merece atenção, visto que quanto maior o número de dígitos, mais informações vão sendo agregadas à divisão terrotorial. Por exemplo, o prefixo de dois dígitos, '35', que consta em todos os municípios do Estado de São Paulo, representa justamente a divisão estadual. Ou seja, conforme se aumenta o número de dígitos, se amplia o detalhamento da divisão territorial, iniciando em **_Estado_**, passando para **_Mesoregião_**, **_Microregião_** e, por fim, **_Município_** (com 7 dígitos). Há ainda maiores detalhamentos da divisão territorial, que não serão abrangidas no presente escopo, que tratam de **_Distritos_** e **_setores censitários_**.

Contudo, como na presenta análise estamos manejando o banco de dados do Ministério da Saúde, vinculado ao [DataSus](https://datasus.saude.gov.br/), pode-se notar que o código dos municípios deles apresenta apenas 6 dígitos. Mais informações sobre isso foram obtidas nesse [documento sobre as bases territoriais utilizadas pelo DataSus](ftp://ftp.datasus.gov.br/territorio/doc/bases_territoriais.pdf). Diante disso foram feitas amostras para se avaliar se os códigos são correspondentes, que resultaram que é correto considerar os códigos equivalentes, sendo que o sétimo dígito corresponde a um código verificador apenas.

Ainda, tal informação foi validada com [um comentário](https://www.guj.com.br/t/resolvido-onde-baixar-a-tabela-com-o-codigo-do-ibge-atualizada-com-os-7-numeros/12014) que se encontrou em um fórum de discussão na internet de 2009, onde determinado usuário mencionou que *"o código antigo tinha 6 números, agora o novo é com sete"*.
